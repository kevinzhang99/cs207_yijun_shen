```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = np.linspace(-1,1,500)
y1 = x**2
y2 = np.sin(x)
y3 = np.cos(x)
plt.plot(x, y1, lw=4, ls='-', label="x^2")
plt.plot(x, y2, lw=4, ls='--', label="sin(x)")
plt.plot(x, y3, lw=4, ls='-', label="log(x)")
plt.legend()
plt.xlabel(r'$x$', fontsize=24)
plt.ylabel(r'$y$', fontsize=24)
plt.title('Three functions for Exercise 1', fontsize=20)
# plt.show()
```

```python
N = 1000
random = np.random.randn(N)
plt.hist(random)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$frequency$', fontsize=20)
```
